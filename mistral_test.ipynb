```python
import os
import json
from datetime import datetime
from dotenv import load_dotenv
from mistralai.client import MistralClient
from mistralai.models.chat_completion import ChatMessage

# Load environment variables
load_dotenv()

# Get API key with error handling
api_key = os.getenv('MISTRAL_API_KEY')
if not api_key:
    raise ValueError("MISTRAL_API_KEY not found in environment variables")

try:
    # Initialize Mistral client
    client = MistralClient(api_key=api_key)
    
    # Use the latest stable model
    model = "pixtral-large-2411"
    
    article_content = """
   陽明(2609)董事會針對現行營運船隊汰舊換新與業務推展需要，決議展開新世代船舶布建計畫，以取得至多13艘8,000 TEU至15,000 TEU級全貨櫃船為範疇，依內部程序辦理後續作業。

陽明表示，以提供客戶完整及高效率全球運輸服務為核心，因此，持續優化航線布局並致力於提升船隊競爭力。本次應中長期業務穩健發展需要，陸續展開新世代節能船舶布建規畫，汰換現行營運中船齡超過20年5,500 TEU至6,500 TEU級距貨櫃船，以強化核心航線營運優勢。

為呼應氣候變遷行動，新船布建將有助於陽明海運未來得自主配合法規與技術條件執行減碳規畫，透過綜合評估節能技術發展及市場潔淨能源取得實務作為，落實逐步減少溫室氣體排放、強化企業經營韌性、提供全球客戶更環保運送服務目標。

此外，陽明積極開發新市場，鑑於印度強勁的經濟表現且為全球快速成長的大型經濟體之一，陽明將新闢印度-歐洲航線（ISE），提供往返印度、斯里蘭卡及巴基斯坦至西北歐各港的貨櫃直航運輸服務，以深耕印巴次大陸往返歐洲市場。

ISE航線將於2025年2月5日啟航，將繞行好望角，靠港順序：喀拉蚩Karachi-哈吉拉Hazira-孟德拉Mundra-那瓦舍瓦Nhava Sheva-可倫坡Colombo-倫敦門戶港London Gateway-鹿特丹Rotterdam-漢堡Hamburg-安特衛普Antwerp-喀拉蚩Karachi。
    """
    
    # 修改範例格式，將相同概念和理由的股票群組在一起
    kv_pattern="""[
    {
        "股票": ["華航", "長榮航"],
        "概念": ["航空", "貨運需求", "成本控管"],
        "理由": "航空業受惠貨運需求強勁及成本控管良好"
    },
    {
        "股票": ["台光電", "台燿"],
        "概念": ["CCL", "AI伺服器", "高階市場"],
        "理由": "高階電子元件市場前景看好"
    }
]"""
    
    # 改進的 prompt
    messages = [
        ChatMessage(
            role="user", 
            content=(
                "你是一位專業的股票分析師，請分析以下文章並依照相同概念和理由來群組股票：\n\n"
                f"{article_content}\n\n"
                "分析規則：\n"
                "1. 將具有相同概念和投資理由的股票群組在一起\n"
                "2. 每個群組必須包含：\n"
                "   - 股票：股票名稱陣列（相同概念和理由的股票列表）\n"
                "   - 概念：1-3個關鍵題材（每個最多8字）\n"
                "   - 理由：分析原因（最多150字）\n"
                "3. 群組原則：\n"
                "   - 相同產業且投資理由相近的股票應該群組在一起\n"
                "   - 即使股票代號相近，如果概念或理由不同也應分開群組\n"
                "   - 確保每個群組的概念和理由都具有獨特性\n"
                "4. 直接輸出JSON陣列格式，不需其他說明文字\n"
                f"5. 參考格式：{kv_pattern}\n"
                "6. 確保完整收錄文章中所有相關股票，但必須依據概念和理由正確群組"
            )
        )
    ]

    # Make the API call
    chat_response = client.chat(
        model=model,
        messages=messages,
        temperature=0.7
    )
    
    # Get the response content
    response_content = chat_response.choices[0].message.content.strip()
    
    # 建立輸出目錄（如果不存在）
    output_dir = "stock_analysis"
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)
    
    # 使用當前時間建立檔案名稱
    current_time = datetime.now().strftime("%Y%m%d_%H%M%S")
    output_file = os.path.join(output_dir, f"stock_analysis_{current_time}.json")
    
    # 將回應內容儲存為 JSON 檔案
    with open(output_file, "w", encoding="utf-8") as f:
        f.write(response_content)
    
    print(f"Analysis has been saved to: {output_file}")

except Exception as e:
    print(f"An error occurred: {str(e)}")
```
