```python
import git

def search_commits_by_keyword(repo_path, keyword):
    repo = git.Repo(repo_path)
    commits = []
    for commit in repo.iter_commits():
        if keyword in commit.message:
            commits.append(commit.hexsha)
    return commits

def get_modified_files_with_extension(commit_hash, extension):
    repo = git.Repo(".")
    commit = repo.commit(commit_hash)
    modified_files = [f for f in commit.stats.files.keys() if f.endswith(extension)]
    return modified_files

repo_path = "."  # 替換為你的Git存儲庫路徑
keyword = "Email"  # 替換為你要搜尋的關鍵字
extension = ".js"  # 替換為你要篩選的附檔名
commit_hashes = search_commits_by_keyword(repo_path, keyword)

print("Commits containing the keyword:")
for commit_hash in commit_hashes:
    modified_files = get_modified_files_with_extension(commit_hash, extension)
    print(commit_hash)
    if modified_files:
        for file in modified_files:
            if file.find('case') > 0:
                print(file)
        print("-" * 50)
```
